```
### create blank white image
import numpy as np
img = np.ones((512,512,3),np.uint8)*255
plt.imshow(img)
plt.show
```

```
## Draw a line on image
img = np.ones((512,512,3),np.uint8)*255
cv2.line(img,(30,300),(400,55),(255,135,0),3)     ##(img,pt1,pt2,color(RGB),width)
plt.imshow(img)
plt.show
```

```
## Draw a rectangle on image
img = np.ones((512,512,3),np.uint8)*255
cv2.rectangle(img,(30,300),(400,55),(135,135,0),3)     ##(img,pt1,pt2,color(RGB),width)
plt.imshow(img)
plt.show
```

```
## Draw a rectangle on image
img = np.ones((512,512,3),np.uint8)*255
cv2.rectangle(img,(30,300),(400,55),(135,25,200),-1)     ##(img,pt1,pt2,color(RGB),width)
plt.imshow(img)
plt.show
```

```
## Draw a rectangle and line on image
img = np.ones((512,512,3),np.uint8)*255
cv2.rectangle(img,(30,300),(400,55),(135,25,200),-1)   ##(img,pt1,pt2,color(RGB),width)
cv2.line(img,(30,300),(400,55),(255,135,0),3)
plt.imshow(img)
plt.show
```

```
## Draw a circle on image
img = np.ones((512,512,3),np.uint8)*255
cv2.circle(img,(300,300),(125),(135,25,200),-1)     ##(img,centre pt, radius,color(RGB),width)
plt.imshow(img)
plt.show
```

```
## Draw an ellipse on image
img = np.ones((512,512,3),np.uint8)*255
cv2.ellipse(img,(300,300),(200,50),0,0,360,(135,25,200),-1)     ##(img,centre pt, axes, angle, start,end,color(RGB),width)
plt.imshow(img)
plt.show
```

```
## Draw an polygon on image
img = np.ones((512,512,3),np.uint8)*255
points = np.array([[10,200],[100,100],[400,200]])
cv2.polylines(img,[points], True,(135,25,200),5)     ##(img,[Points], isclosed,color(RGB),width)
plt.imshow(img)
plt.show
```

```
## Write Text on image
img = np.ones((512,512,3),np.uint8)*255
cv2.putText(img,"Nikhil",(100,200),cv2.FONT_HERSHEY_DUPLEX,3,(135,25,200),5)     ##(img,"TExt",position, font styles,color(RGB),width)
plt.imshow(img)
plt.show
```

```
##Image blending or Transparency or watermark
img1 = cv2.imread('/content/Nik 3.5x4.5.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img1 = cv2.resize(img1, (512,512))
plt.imshow(img1)
plt.show
```

```
## Write Text on image
img2 = np.ones((512,512,3),np.uint8)*255
cv2.putText(img2,"Nikhil",(100,400),cv2.FONT_HERSHEY_DUPLEX,3,(255,25,2),5)     ##(img,"TExt",position, font styles,color(RGB),width)
plt.imshow(img2)
plt.show
img2.shape
```

```
final_image = cv2.addWeighted(img1,0.80,img2,0.20,0)
final_image = cv2.cvtColor(final_image, cv2.COLOR_BGR2RGB)
cv2.imwrite("My_image.jpg",final_image)       ## to save
plt.imshow(final_image)
plt.show
```
